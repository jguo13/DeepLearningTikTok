Now, we will take screen captures of the videos, converting them from video files to image files at a rate of 5 frames per second. The videos have been preloaded into the deeplearn folder in my google drive, and they are videos that include Trump and some that do not gathered from the tiktok api. We repeat the loop 18 times for 18 different tiktok videos.

The code to capture the video can be studied futher at this link: https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_gui/py_video_display/py_video_display.html

```
#import libraries for converting the videos to images
%matplotlib inline
import cv2  
import math
import matplotlib.pyplot as plt
import pandas as pd
from keras.preprocessing import image
import numpy as np
from keras.utils import np_utils
from skimage.transform import resize
 
```

```
count = 0
videoFile = "trump_biden.mp4" #name of tiktok video
cap = cv2.VideoCapture(videoFile)   # capturing the video from the given path
frameRate = cap.get(5) #frame rate
x=1
while(cap.isOpened()):
    frameId = cap.get(1) #we get the current frame
    ret, frame = cap.read() #returns True/False. If frame is read correctly it will return True. If it returns False you are at the end of the video.
    if (ret != True):
        break #This means you are at the end of the video so break out of while loop
    if (frameId % math.floor(frameRate) == 0):
        filename ="frame%d.jpg" % count;count+=1
        cv2.imwrite(filename, frame) #saves the last frame in the 5 frames/second
cap.release()
print ("Done!")
```

The videos have now been broken up into image files. We will train our classifier to label 1 if Trump is detected in an image, and 2 if no Trump is detected in the image. After doing so, I named the csv "mapping_trumpbiden_final.csv" and upload it to the google drive.
```
data = pd.read_csv('mapping_trumpbiden_final.csv')     # reading the csv file
data.head()      # printing first five rows of the file
```

We now convert all the images into arrays and store in variable X in order to train the model on them.

```
X = [ ]     # creating an empty array
for img_name in data.Image_ID:
    img = plt.imread('' + img_name)
    X.append(img)  # storing each image in array X
X = np.array(X)    # converting list to array
print(X)
```

```
y = data.Class #set the labels
dummy_y = np_utils.to_categorical(y)    # one hot encoding Classes, we need this to match output shape of model
print(dummy_y)
```

We now reshape the images into arrays of size (224,224). This is because we train the images on the vgg16 model, which expects those dimensions for the input images.

```
image = []
for i in range(0,X.shape[0]):
    a = resize(X[i], preserve_range=True, output_shape=(224,224)).astype(int)      # reshaping to 224*224*3
    image.append(a)
X = np.array(image)
```

```
from keras.applications.vgg16 import preprocess_input
X = preprocess_input(X)  
print(X.shape)
```

Break the data down into training and validation split. 

```
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, dummy_y, test_size=0.3, random_state=42)    # preparing the validation set
```

Now we feed the images through the vgg16 covulutional network to output the features

```
from keras.models import Sequential
from keras.applications.vgg16 import VGG16
from keras.layers import Dense, InputLayer, Dropout, Flatten
```

```
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))    # include_top=False to remove the top layer
```

```
X_train = base_model.predict(X_train)
X_valid = base_model.predict(X_valid)
X_train.shape, X_valid.shape
```

Here, we start to build our own model to recognize Trump in tiktok videos. The important part is that the input layer should match the shape of your data, and the output shape should match the size of your labels. If your input shape has more dimensions than your output shape, be sure to add a Flatten() layer in between to reduce the dimensions.

```
#Building the model
model = Sequential()
model.add(InputLayer((7,7,512,)))    # input layer: ensure this is the same size as your data
model.add(Flatten())
model.add(Dense(units=1024, activation='sigmoid'))   # hidden layer: these layers are somewhat flexible so you can change these as you see fit
model.add(Dropout(0.5))      # adding dropout prevents overfitting
model.add(Dense(units=512, activation='sigmoid'))    # hidden layer
model.add(Dropout(0.5))      # adding dropout
model.add(Dense(units=256, activation='sigmoid'))    # hidden layer
model.add(Dropout(0.5))      # adding dropout
model.add(Dense(3, activation='softmax'))            # output layer: ensure this is the same size as your label data
```

```
model.summary()
```

```
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```
print(X_train.shape)
print(y_train.shape)
print(y_valid.shape)
```

Since we hand labeled all the images and since that takes an incredible amount of time, we can now supplement our work with data augmention. In the below code, we create an ImageDataGenerator that we use to fit our model with. The ImageDataGenerator will return augmented images from your original input data in batches. 

```
from tensorflow.keras.preprocessing.image import ImageDataGenerator
datagen = ImageDataGenerator(        
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True)
datagen.fit(X_train)
```

```
model.fit(datagen.flow(X_train, y_train, batch_size=32), steps_per_epoch=len(X_train) / 32, epochs = 50, validation_data = datagen.flow(X_valid, y_valid, batch_size=32))
```
